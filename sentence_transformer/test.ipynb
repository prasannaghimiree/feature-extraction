```python
import matplotlib.pyplot as plt
import numpy as np

# Data
epochs = [1,2,3,4,5,6,7,8,9]

train_acc = [0.67905,0.747259,0.74056,0.763094,0.777101,0.755177,0.778928,0.795981,0.797808]
val_acc   = [0.63388,0.661202,0.68306,0.677596,0.688525,0.699454,0.715847,0.693989,0.693989]

# Plot
plt.figure(figsize=(12,6))

plt.plot(epochs, train_acc, marker='o', label='Train Accuracy')
plt.plot(epochs, val_acc, marker='o', label='Validation Accuracy')

# FULL SCALE (no zoom)
plt.ylim(0.5, 1)
plt.yticks(np.arange(0.5, 0.9, 0.025))

plt.title("Training vs Validation Accuracy (Full Scale View)")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
